# Violin Plots
Violin plots are another visual aid for visual exploratory data analysis. 

```python
sns.set_style("whitegrid")
sns.catplot(x="day", y="tip", hue="sex", kind="violin", split=True, height=10, aspect=11.7/8.27, data=tips);
```
